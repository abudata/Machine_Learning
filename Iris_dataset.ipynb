```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

The iris object returned by load_iris is a Bunch object, which is very similar to a dictionary i.e., it contains keys and values. 
In order to check the keys of this dataset, we can use:

```python
print('Keys of iris_dataset: \n {}'.format(iris_dataset.keys()))
```

The value of the key 'DESCR'is a short description of the dataset. We can see it's value as below:

```python
print(iris_dataset['DESCR'][:193] + '\n ...')
```

Also, we can look at the values of 'target_names' as well as 'feature_names'

```python
print('Target names:\n {}'.format(iris_dataset['target_names']))
print('========================================================')
print('Feature names:\n {}'.format(iris_dataset['feature_names']))
```

The entire data can be found in the 'data' and 'target' fields. data itself contains the numeric values of each 'target_names'.

```python
print('Type of data is: \n {}'.format(type(iris_dataset['data'])))
print(30 *'==')
print('Shape of data is: \n {}'.format(iris_dataset['data'].shape))
```

data fields contains 150 measurements of differetn flowers with 4 feauters.
It is important to mention as well that the individual items (rows) are called 'samples' in machine learning. Also, their properties (columns) are called 'features'

```python
print('First ten samples of data is: \n {}'.format(iris_dataset['data'][:10]))
```

There are four columns for this data which corresponds to 'sepal length', 'sepal width, 'petal length' and 'petal width' respectively with the units of 'cm'. So, the first five flowers have a petal width of 0.2 cm.

```python
print('Feature names:\n {}'.format(iris_dataset['feature_names']))
```

It's worth mentioning again that the 'target' field contains the species of the flowers with a NmuPy array data type.

```python
print('Type of target: {}'.format(type(iris_dataset['target'])))
```

```python
print('Shape of target is: {}'.format(iris_dataset['target'].shape))
```

The species are encoded as integers from 0 to 2 as follows:
